```python
PROJECT_PATH = '/content/drive/MyDrive/2022CocacolaDrive/'
IMG_PATH = f'{PROJECT_PATH}MonoDTR/data/KITTI/testing/image_2/'
CALIB_PATH = f'{PROJECT_PATH}MonoDTR/data/KITTI/testing/calib/'

CASE_DIV = 'case10'
RAW_IMG_PATH_A = f'{PROJECT_PATH}팀문서/3. 테스트 영상/2차 테스트 영상_20221113/{CASE_DIV}/a/'
RAW_IMG_PATH_B = f'{PROJECT_PATH}팀문서/3. 테스트 영상/2차 테스트 영상_20221113/{CASE_DIV}/b/'
# RAW_IMG_NAME = 'timea0900.png'

RAW_CALIB_PATH_A = f'{PROJECT_PATH}MonoDTR/data/KITTI/testing/calib_bu/A.txt'
RAW_CALIB_PATH_B = f'{PROJECT_PATH}MonoDTR/data/KITTI/testing/calib_bu/B.txt'
```

1. 이미지 불러오기
2. 이미지 리사이징
3. 이미지 저장
4. 캘리브레이션 복사

```python
a_index = 1319
b_index = 544
end_a_index = 1559

set_size = end_a_index-a_index
set_sec = round(set_size/30, 1)

#########################
indexes = [a_index, b_index]

prg_N = set_size * 2
n = 1

for carNum, index in enumerate(indexes):
    for i in range(index, index + set_size):

        if carNum == 0:
            src_name = RAW_IMG_PATH_A+CASE_DIV+'a'+"%04d" % i+'.png'
            
            res_name = "%06d" % (i-index)
        else:
            src_name = RAW_IMG_PATH_B+CASE_DIV+'b'+"%04d" % i+'.png'
            res_name = "%06d" % (set_size+i-index)

        IMG_PATH = PROJECT_PATH+'MonoDTR/data/KITTI/testing/image_2/'
        CALIB_PATH = PROJECT_PATH+'MonoDTR/data/KITTI/testing/calib/'
        IMG_PATH = IMG_PATH + res_name + '.png'
        CALIB_PATH = CALIB_PATH + res_name + '.txt'

        ### LOG
        printout = f'\r{src_name} : {n}/{prg_N}'
        stdout.write(printout)
        n += 1
        ###

        kittiwidth = 1242
        kittiheight = 375

        # 이미지 로드
        # print(src_name)
        image = cv2.imread(src_name, cv2.IMREAD_COLOR)
        rgb_image = image.transpose(0, 1, 2)
        height = rgb_image.shape[0]
        width = rgb_image.shape[1]

        # 이미지 리사이징
        aspect_ratio = float(kittiwidth) / width
        dsize = (width, int(height * aspect_ratio))
        resized = cv2.resize(rgb_image, dsize, interpolation=cv2.INTER_AREA)

        # 이미지 저장
        cv2.imwrite(IMG_PATH, resized, [cv2.IMWRITE_JPEG_QUALITY, 50])

        # 캘리브레이션 복사
        if carNum == 0:
            shutil.copy(RAW_CALIB_PATH_A, CALIB_PATH)
        else:
            shutil.copy(RAW_CALIB_PATH_B, CALIB_PATH)

print(f'\n{set_size} / {IMG_PATH}')
```
